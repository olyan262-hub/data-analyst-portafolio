```python
import pandas as pd

df = pd.read_csv("tickets.csv")
```

```python
# Total tickets
len(df)
```

```python
# Average response time
df["response_time_hours"].mean()
```

```python
# Tickets by category
df.groupby("category")["ticket_id"].count().sort_values(ascending=False)
```

```python
# Average response time by priority
df.groupby("priority")["response_time_hours"].mean().sort_values(ascending=False)
```

```python
tickets_by_category = (
    df.groupby("category")["ticket_id"]
    .count()
    .sort_values(ascending=False)
)
tickets_by_category
```

```python
# High delay tickets (>8 hours)
df[df["response_time_hours"] > 8]
```

```python
import matplotlib.pyplot as plt

tickets_by_category.plot(kind="bar")
plt.title("Tickets by Category")
plt.xlabel("Category")
plt.ylabel("Number of Tickets")
plt.show()
```
